DROP EXTRA AND LESS USEFUL COLUMN

```python
df.drop('taster_twitter_handle', axis=1, inplace=True, errors='ignore')
```

```python
# Fill missing 'country' values with the mode
country_mode = df['country'].mode()[0]
df['country'].fillna(country_mode, inplace=True)

# Verify null counts for 'country'
df['country'].isnull().sum()
```

```python
# Fill missing 'price' values with the median
price_median = df['price'].median()
df['price'].fillna(price_median, inplace=True)

# Verify null counts for 'price'
df['price'].isnull().sum()
```

```python
# Fill missing 'winery' values with the mode
winery_mode = df['winery'].mode()[0]
df['winery'].fillna(winery_mode, inplace=True)

# Verify null counts for 'winery'
df['winery'].isnull().sum()
```

```python
# Fill missing 'province' values with the mode
province_mode = df['province'].mode()[0]
df['province'].fillna(province_mode, inplace=True)

# Verify null counts for 'province'
df['province'].isnull().sum()
```
